```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("../data/vehicles.csv")
```

## Histograma del precio

```python
plt.figure(figsize=(8,5))
plt.hist(df['price'], bins=50, color='skyblue', edgecolor='black')
plt.title('Distribución de precios')
plt.xlabel('Precio ($)')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```

## Boxplot de precio por condición

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='condition', y='price', data=df)
plt.title('Precio por estado del vehículo')
plt.xlabel('Condición')
plt.ylabel('Precio ($)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

## Gráfico de dispersión año vs precio

```python
plt.figure(figsize=(8,5))
plt.scatter(df['year'], df['price'], alpha=0.5)
plt.title('Relación entre año de fabricación y precio')
plt.xlabel('Año')
plt.ylabel('Precio ($)')
plt.grid(True)
plt.show()
```

## Gráfico de dispersión KM vs precio

```python
plt.figure(figsize=(8,5))
plt.scatter(df['KM'], df['price'], alpha=0.5, color='orange')
plt.title('Relación entre kilometraje y precio')
plt.xlabel('Kilometraje (KM)')
plt.ylabel('Precio ($)')
plt.grid(True)
plt.show()
```

## Matriz de correlación (Heatmap)

```python
plt.figure(figsize=(10,8))
correlaciones = df.corr(numeric_only=True)
sns.heatmap(correlaciones, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de correlación')
plt.show()
```
